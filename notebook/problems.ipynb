## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
PSUEDOCODE

** make a counter of how many times we want to run the experiement 

create 2 dice objects that store nothing at the start
use the random function from (1-6) for each dice object
    
create a third object for the sum of both dice

create a if loop,
if sum greater than 7, or an even number (use modulo % 2 == 0), then add 1 to a new counter object

make rate object as previous number divided by 1000
```

```python
import numpy as np
import random

total_iterations = 1000
dice1 = 0
dice2 = 0
counter = 0
for count in range(total_iterations):
    dice1 = random.randint(1,6)
    dice2= random.randint(1,6)
    sum = dice1 + dice2

    if sum > 7:
        counter = counter + 1
    elif sum % 2 == 0:
        counter = counter + 1
    else:
        pass

rate = counter / 1000

print(rate)
```
